# ***DATA CLEANING***

  **Droping the features which do not effect the target**

1.   *' PassengerId '*
2.   *' Name'*

1.   *' Ticket '*

1.   *' Cabin '*













```python
titanic=titanic.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```



**Making all the features into int or float.**



```python
titanic["Sex"]=titanic["Sex"].map({"male":0,"female":1})
titanic["Embarked"]=titanic["Embarked"].map({"S":1,"C":2,"Q":3})
```



**Checking for any empty/NaN.**



```python
titanic.isnull().sum()
```

*  **Conclusion**

   *   *'Age'* has 177 empty spaces

    *   filling the empty spaces with mean value of *'Age'*.
   *   *'Embarked'* has 2 empty spaces
    *   as *'Embarked'* is a place we cannot use mean, instead we can use mode of *'Embarked'*.






```python
titanic["Age"].fillna(value=titanic["Age"].mean(),inplace=True)
titanic["Embarked"].fillna(value=titanic["Embarked"].mode()[0],inplace=True)
```

```python
titanic.isnull().sum()
```

# ***SPLITTING THE DATA INTO INPUT(X) AND TARGET(Y)***

**Drop the *'Survived'* coloumn and make the remaining dataframe as input(x).**
<br>
**Make *'Survived'* as target(y).**



```python
x=titanic.drop(['Survived'],axis=1)
y=titanic['Survived']
```

```python
x.head()
```

```python
y.head()
```

**Splitting data int train set and test set**
  *   make 80% data as train set (x_train, y_train).
  *   make 20% data as test set (x_test, y_test).





```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=0)
```

# ***STANDARDIZATION USING STANDARDSCALER***

```python
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```


# ***TRAINING THE MODEL***
**Training the 'model' using LogisticRegression with train set (x_train, y_train).**



```python
lg=LogisticRegression()
model=lg.fit(x_train,y_train)
```



**Checking the accurcay of our model using accuracy_score.**



```python
predict=model.predict(x_train)
accuracy_score(y_train,predict)
```

```python
y_predict=model.predict(x_test)
accuracy_score(y_test,y_predict)
```
